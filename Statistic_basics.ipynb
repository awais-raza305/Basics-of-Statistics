Q1. Types of Data and Scales
Qualitative (Categorical) Data: Describes categories or groups.
Examples: Gender (Male/Female), Eye color (Blue, Brown).

Quantitative (Numerical) Data: Represents numbers and measurable quantities.
Examples: Height (cm), Age (years).

Scales:

Nominal: Categories without order (e.g., blood type: A, B, AB, O).

Ordinal: Categories with a meaningful order (e.g., survey ratings: Poor, Fair, Good).

Interval: Numerical scale with equal intervals but no true zero (e.g., temperature in Celsius).

Ratio: Interval scale with a true zero point (e.g., weight, height).

Q2. Measures of Central Tendency
Mean: Average of data. Use when data is symmetric and no extreme outliers.

Median: Middle value. Use when data is skewed or has outliers.

Mode: Most frequent value. Useful for categorical data or to find the most common item.

Q3. Concept of Dispersion; Variance and Standard Deviation
Dispersion: Measures the spread or variability in data.

Variance: Average of squared differences from the mean; indicates spread.

Standard Deviation: Square root of variance; expressed in same units as data; easier to interpret.

Q4. What is a Box Plot?
A graphical summary of data showing median, quartiles, and potential outliers.

Tells you about data distribution, central tendency, spread, and outliers visually.

Q5. Role of Random Sampling
Ensures each member of a population has an equal chance of selection.

Helps produce unbiased samples that represent the population.

Allows valid inferences and generalizations about the whole population.

Q6. Skewness and Its Types
Skewness: Measures asymmetry in the data distribution.

Types:

Positive skew (right tail longer, mean > median)

Negative skew (left tail longer, mean < median)

Symmetric (mean ≈ median)

Effect: Skewness affects the choice of statistical methods and interpretation, especially mean and median.

Q7. Interquartile Range (IQR) and Outliers
IQR: Difference between Q3 (75th percentile) and Q1 (25th percentile).

Use: Identifies spread of middle 50% of data; detects outliers if data points lie beyond 1.5 × IQR above Q3 or below Q1.

Q8. Conditions for Binomial Distribution
Fixed number of trials (n).

Each trial has two outcomes: success or failure.

Probability of success (p) is constant.

Trials are independent.

Q9. Normal Distribution and Empirical Rule
Properties: Symmetric, bell-shaped, mean = median = mode, defined by mean (μ) and standard deviation (σ).

Empirical Rule:

68% of data within 1σ of mean

95% within 2σ

99.7% within 3σ

Q11. Random Variable: Discrete vs Continuous
Random Variable: A variable whose values are outcomes of a random phenomenon.

Discrete: Countable outcomes (e.g., number of heads in coin toss).

Continuous: Infinite possible values within a range (e.g., height, time).

Q12. Example Dataset, Covariance, and Correlation
Dataset:
X: [2, 4, 6, 8, 10]
Y: [5, 9, 12, 15, 20]

Covariance: Measures joint variability.

Correlation: Standardized covariance between -1 and 1, shows strength/direction of linear relationship.
